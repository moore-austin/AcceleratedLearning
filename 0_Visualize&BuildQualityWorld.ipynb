```python
def find_neighbors(number_six, distance):
  neighbor_list = []
  
  x_num = number_six[0:2]
  y_num = number_six[2:4]
  z_num = number_six[4:6]

  xyz_vectors = [x_num, y_num, z_num]

  xyzs = []

  # Generate list of lists containing all vectors
  for vector in xyz_vectors:
    vectors = iterate_distance(vector, distance)
    
    xyzs.append(vectors)

  # Generate list containing all iteratons of this list, which are the neighbors
    
  # Constructing First Two Nums
  iterator = 0

  points_2d = []

  for vector in xyzs[0]:
    while iterator < len(xyzs[0]):
      points_2d.append(vector + (xyzs[1][iterator]).zfill(2))
      iterator = iterator + 1

    iterator = 0

  len(points_2d)

  # Adding Third Num to Construction
  iterator = 0
  points_3d = []

  for vector in points_2d:
    while iterator < len(xyzs[2]):
      points_3d.append(vector + (xyzs[2][iterator]).zfill(2))
      iterator = iterator + 1

    iterator = 0

  # Removing All Out of Boundary Node Spaces in Cube
  true_node_spaces = []

  ## Remove Values Too High (ex. 100 in num set)
  for node in points_3d:
    if len(node) <= 6:
      true_node_spaces.append(node)

  ## Removing All Values Too Low (ex. -1 in num set)
  for node in true_node_spaces:
    match = re.search('-', node)
    if match:
      true_node_spaces.remove(node)

  return true_node_spaces



```

```python
def iterate_distance(num, distance):
  vectors = []


  iterator = 0

  while iterator <= distance:

    # Positives
    interval = int(num) + iterator
    vectors.append(str(interval))

    # Negatives
    interval = int(num) - iterator
    vectors.append(str(interval))

    iterator = iterator + 1

  return vectors
```

```python
def get_10k_node_file(number):
  # Break Apart Input Num
  first_num = str(number[0] + number[1])
  second_num = str(number[2] + number[3])


  # Get Folder Location From 10k Number Input
  world_path = '/content/drive/MyDrive/QualityWorldPalaceData'
  node_folders = os.listdir(world_path)


  # Get Quality Drive 100 System Folders
  qd = '/content/drive/MyDrive' #qd = QualityDrive
  nf = [os.path.join(qd, f) for f in os.listdir(qd)
    if os.path.isdir(os.path.join(qd,f))] # nf = list of NodeFolders in QualityDrive
  animalFolderRegex = re.compile('.*?\d*\^\<\w.*') # Put the regex object in a variable
  af = [folder for folder in nf if animalFolderRegex.match(folder)] # Sorts for AFolders

  # Get value_object_image
  ## Sort Via Input to Get Folder
  regex_input_number = re.compile(first_num + '-' + second_num)
  folder_result = [folder for folder in node_folders if regex_input_number.match(folder)]
  node_path = Path(world_path) / Path(folder_result[0])

  return node_path
```

```python
def get_node_files(number): # Get's six digit node file, if there is one
  file_result = ''

  first_num = str(number[0] + number[1])
  second_num = str(number[2] + number[3])
  third_num = str(number[4] + number[5])

  # Get Folder Location From 10k Number Input
  world_path = '/content/drive/MyDrive/QualityWorldPalaceData'
  node_folders = os.listdir(world_path)

  ## Sort Via Input to Get Folder
  regex_input_number = re.compile(first_num + '-' + second_num)
  folder_result = [folder for folder in node_folders if regex_input_number.match(folder)]
  node_path = Path(world_path) / Path(folder_result[0])

  # Get files if any
  node_files = os.listdir(node_path)
  regex_input_number = re.compile('^' + third_num)

  file_results = [node for node in node_files if regex_input_number.match(node)]

  # Adding Path to File
  file_paths = []

  for file in file_results:
    file = node_path / Path(file)
    file_paths.append(file)

  return file_paths
```

```python
def find_uid_file(number):
  # Get Folder
  node_folder = get_10k_node_file(number)
  # Search for UID FIle in Folder
  node_files = os.listdir(node_folder)
  for file in node_files:
    match = re.search('UID-*|', file)
    if match:
      file_name = file
      uid_file = node_folder / Path(file_name)
  return uid_file
```

```python
# Create offset_grid and return coordinates
def get_mesh(lat, lon, dist, coors):
  # Calculate - and max range over an axis to create grid (both directions)
  mini, maxi = -dist*coors, dist*coors

  # Calculate number of points over axis, both ways + center point
  n_coord = coors*2+1

  # Create an axis from min to max value with required number of coordinates
  axis = np.linspace(mini, maxi, n_coord)

  # Create an "offset_grid" for X and Y values for both axis
  X, Y = np.meshgrid(axis, axis)


  # Calculate offset foordinates for "offset_grid" in meters
  mesh = lat_long_offset_vec(lat, lon, X, Y)

  # Transpose to get the (x,y) values for the offset_grid's shape
  mesh_x_y_format = np.stack(mesh).transpose(1, 2, 0)

  return mesh_x_y_format

```

```python
# Function to offset lat, long, with x & y in meters
def lat_long_offset(lat, lon, x, y):
    '''
    Note from creator (https://stackoverflow.com/questions/69409596/how-can-i-create-grid-of-coordinates-from-a-center-point-in-python)
    lat, lon : Provide lat lon coordinates
    x, y : Provide offset of x and y on lat and long respectively
           This needs to be in meters!
           
    The approximation is taken from an aviation formula from this stack exchange 
    https://gis.stackexchange.com/questions/2 pm 951/algorithm-for-offsetting-a-latitude-longitude-by-some-amount-of-meters
    '''
    # Earth's radius, sphere
    R=6378137

    # Coordinate offsets in radians
    dLat = x/R
    dLon = y/(R*np.cos(np.pi*lat/180))

    # OffsetPosition, decimal degrees
    latO = lat + dLat * 180/np.pi
    lonO = lon + dLon * 180/np.pi

    return latO, lonO

# Create a vectorized offset function, pushes geo input through np.vectorize
lat_long_offset_vec = np.vectorize(lat_long_offset)
```

```python
from pandas.tseries.offsets import Second
def nearest_neighbors(number, loci_distance=4):
  '''
  Takes in list of connections.

  Loci_distance is the number of locis away from node on map.
  
  Gets nearest neighbor nodes of number entered into function
  and returns a list of connection coordinates.
  '''
  nearest_neighbors = []

  # Breaking Num Apart to Manipulate
  first_animal_num = number[0] + number[1]
  second_animal_num = number[2] + number[3]

  first = int(number[0])
  second = int(number[1])
  third = int(number[2])
  fourth = int(number[3])

  neighbor_nums = [first, second, third, fourth]
  
  # Generating nearest_neighbor list
  count = 1

  # For Higher Neighbors
  while count != loci_distance:
    if int(first_animal_num) < 99:
      # Add 1 to first num, then append to list
      first_num = int(first_animal_num)
      first_animal_num = (str(first_num + 1)).zfill(2)
      nearest_neighbors.append(first_animal_num + second_animal_num) # 1 added to first_num only

    if int(second_animal_num) < 99:
    # Add one to second num, then add to list
      if int(first_animal_num) == 99: # To account for this variable being skipped over during '99' input
        first_num = first_animal_num

      second_num = int(second_animal_num)
      second_animal_num = (str(second_num + 1)).zfill(2)
      nearest_neighbors.append(str(first_num).zfill(2) + second_animal_num) # 1 added to second_num only

      if int(first_animal_num) < 99: # To account for redundantly adding first_num in again when it's reached it's limit
        nearest_neighbors.append(first_animal_num + second_animal_num) # 1 added to both
      
    count = count + 1

  # Resetting variables
  first_animal_num = number[0] + number[1]
  second_animal_num = number[2] + number[3]
  count = 1

  # Repeating for Lower Neighbors

  while count != loci_distance:
    if int(first_animal_num) > 0:
      # Add 1 to first num, then append to list
      first_num = int(first_animal_num)
      first_animal_num = (str(first_num - 1)).zfill(2)
      nearest_neighbors.append(first_animal_num + second_animal_num) # 1 added to first_num only

    if int(second_animal_num) > 0:
    # Add one to second num, then add to list
      if int(first_animal_num) == 00: # To account for this variable being skipped over during '99' input
        first_num = first_animal_num

      second_num = int(second_animal_num)
      second_animal_num = (str(second_num - 1)).zfill(2)
      nearest_neighbors.append(str(first_num).zfill(2) + second_animal_num) # 1 added to second_num only

      if int(first_animal_num) > 0: # To account for redundantly adding first_num in again when it's reached it's limit
        nearest_neighbors.append(first_animal_num + second_animal_num) # 1 added to both
      
    count = count + 1

  
  return nearest_neighbors
```

```python

# importing google_images_download module

from google_images_download import google_images_download 
 
# creating object

response = google_images_download.googleimagesdownload() 
 
search_queries =[
'The smartphone also features an in display fingerprint sensor.',
'The pop up selfie camera is placed aligning with the rear cameras.',
'''In terms of storage Vivo V15 Pro could offer
   up to 6GB of RAM and 128GB of onboard storage.''',
'The smartphone could be fuelled by a 3 700mAh battery.',
]
 
 

def downloadimages(query):

    # keywords is the search query

    # format is the image file format

    # limit is the number of images to be downloaded

    # print urls is to print the image file url

    # size is the image size which can

    # be specified manually ("large, medium, icon")

    # aspect ratio denotes the height width ratio

    # of images to download. ("tall, square, wide, panoramic")

    arguments = {"keywords": query,

                 "format": "jpg",

                 "limit":4,

                 "print_urls":True,

                 "size": "medium",

                 "aspect_ratio":"panoramic"}

    try:

        response.download(arguments)

     

    # Handling File NotFound Error    

    except FileNotFoundError: 

        arguments = {"keywords": query,

                     "format": "jpg",

                     "limit":4,

                     "print_urls":True, 

                     "size": "medium"}

                      

        # Providing arguments for the searched query

        try:

            # Downloading the photos based

            # on the given arguments

            response.download(arguments) 

        except:

            pass
 
# Driver Code

for query in search_queries:

    downloadimages(query) 

    print() 
```
